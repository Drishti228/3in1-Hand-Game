```python
import random #for getting random values by the computer
#import os     #for clearing the screen when os.clear() is called
import time   # for making it realistic we are going to put sometime delay from the computer side time.sleep() 
from PIL import Image
import requests
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import webbrowser


def rps_rules():
    print()
    print("Rules for Rock-Paper-Scissors : ")
    print()
    print("Rock crushes Scissors")
    print("Scissors cuts Paper")
    print("Paper covers Rock")
    print()
    
def rpsls_rules():
    print()
    print("Rules for Rock-Paper-Scissors-Lizard-Spock : ")
    print()
    print("Scissors cuts Paper")
    print("Paper covers Rock")
    print("Rock crushes Lizard")
    print("Lizard poisons Spock")
    print("Spock smashes Scissors")
    print("Scissors decapitates Lizard")
    print("Lizard eats Paper")
    print("Paper disproves Spock")
    print("Spock vaporizes Rock")
    print("Rock crushes Scissors")
    print()
    

def nmass_rules():
    print()
    print("Rules for Nuke-Snake-Man-Sponge-Alien : ")
    print()
    print("Nuke incinerate Snake")
    print("Snake Bites Man")
    print("Man cleans with Sponge")
    print("Sponge intrigues Alien")
    print("Alien defuses nuke")
    print("Nuke incinerate Man")
    print("Man disproves Alien")
    print("Alien mutates Snake")
    print("Snake swallows Sponge")
    print("Sponge cleans Nuke")
    print()
    
def userwins():
    img = Image.open(requests.get("https://www.meme-arsenal.com/memes/8607903f9cdef626af47110cdb618123.jpg",stream=True).raw)
    imgplot = plt.imshow(img)
    ax = plt.gca()
    ax.axes.xaxis.set_ticks([])
    ax.axes.yaxis.set_ticks([])
    plt.show()

def userloses():
    img = Image.open(requests.get("https://i.imgflip.com/35bnby.jpg",stream=True).raw)
    imgplot = plt.imshow(img)
    ax = plt.gca()
    ax.axes.xaxis.set_ticks([])
    ax.axes.yaxis.set_ticks([])
    plt.show()
    
    
def user_win_rps(player, opponent):
    # return true if player wins
    # r > s, s > p, p > r
    if (player == 'r' and opponent == 's') or (player == 's' and opponent == 'p') \
        or (player == 'p' and opponent == 'r'):
        return True

def user_win_rpsls(player, opponent):
    #return True if player wins
    # s>p,p>r,r>l,l>v,v>s,s>l,l>p,p>v,v>r,r>s
    if(player =='s' and opponent =='p') or (player =='p'and opponent =='r') or (player =='r'and opponent =='l') or \
    (player =='l' and opponent == 'v') or (player == 'v'and opponent == 's') or (player =='s' and opponent =='l') or \
    (player == 'l' and opponent == 'p' ) or (player == 'p' and opponent == 'v' ) or (player == 'v' and opponent == 'r' ) or \
    (player == 'r' and opponent == 's' ):
        return True
    
    
def user_win_nmass(player, opponent):
    #return True if player wins
    # n>s, s>m, m>g, g>a, a>n, n>m, m>a, a>s, s>g, g>n
    if(player =='n' and opponent == 's') or (player == 's'and opponent == 'm') or (player == 'm' and opponent == 'g') or \
    (player =='g' and opponent =='a' ) or (player =='a' and opponent =='n' ) or (player =='n' and opponent =='m' ) or \
    (player =='m' and opponent =='a' ) or (player =='a' and opponent =='s' ) or (player =='s' and opponent =='g' ) or \
    (player =='g' and opponent =='s'):
        return True
    

def rps():
    global name
    
    print("Let's play Rock-Paper-Scissor, the classic.")
    ans= input("If you know how to play, write yes, otheriwse write NO")
        
    if ans.lower()=='no':
        rps_rules()
        #continue
        print("Hope you got that.")
        
    ans1=input('''
    If you want to play then write "PLAY",
    If you want to quit the game then write "EXIT" ''')
        
    if ans1.lower()=='exit':
        exit()
            
    if ans1.lower()=='play':
        print(''' 
        R/r for Rock
        P/p for Paper
        S/s for Scissor''')
        
        user=input("What's your move?")
            
        print("Computer making a move....")
        print()
        time.sleep(4)
            
        computer = random.choice(['r', 'p', 's'])
        print("Computer chooses ", computer.upper())
            
        if user.lower() == computer.lower():
            print("oops,IT'S A TIE!!")
            
        if user_win_rps(user, computer):
            userwins()
            print(name, "WINS!!")
        else:
            return userloses()
            print("Computer WINS!!")
                
    else:
        print("Wrong Input!!")
        rps_rules()
        #continue
                
                
def rpsls():
    global name
    
    print("Let's play Rock-Paper-Scissor-Lizard-Spock, here comes the vulcan salute.")
    ans= input("If you know how to play, move on, otheriwse write NO")
        
    if ans.lower()=='no':
        print("Don't worry bud, got you ;D")
        webbrowser.open('https://www.youtube.com/watch?v=iapcKVn7DdY', new=2) 
        rpsls_rules()
        #continue
        print("Hope you got that.")
        
    ans1=input('''
    If you want to play then write "PLAY",
    If you want to quit the game then write "EXIT" ''')
        
    if ans1.lower()=='exit':
        exit()
            
    if ans1.lower()=='play':
        print(''' 
        R/r for Rock
        P/p for Paper
        S/s for Scissor
        L/l for Lizard
        V/v for Spock''')
        user=input("What's your move?")
        print("Computer making a move....")
        print()
        time.sleep(4)
            
        computer = random.choice(['r', 'p', 's','l','v'])
        print("Computer chooses ", computer.upper())
            
        if user.lower() == computer.lower():
            print("oops,IT'S A TIE!!")
            
        if user_win_rpsls(user, computer):
            userwins()
            print(name, "WINS!!")
        
        else:
            return userloses()
            print("Computer WINS!!")
                
    else:
        print("Wrong Input!!")
        rpsls_rules()
        #continue
                
def nmass():
    global name
    
    print("Let's play Nuke-Snake-Man-Sponge-Alien, be ready to beam up.")
    ans= input("If you know how to play, move on, otheriwse write NO")
        
    if ans.lower()=='no':
        nmass_rules()
        #continue
        print("Hope you got that.")

    ans1=input('''
    If you want to play then write "PLAY",
    If you want to quit the game then write "EXIT" ''')
        
    if ans1.lower()=='exit':
        exit()
            
    if ans1.lower()=='play':
        print(''' 
        N/n for Nuke
        S/s for Snake
        M/m for Man
        G/g for Sponge
        A/a for Alien''')
        user=input("What's your move?")
            
        print("Computer making a move....")
        print()
        time.sleep(4)
            
        computer = random.choice(['n','s','m','g','a'])
        print("Computer chooses ", computer.upper())
            
        if user.lower() == computer.lower():
            print("oops,IT'S A TIE!!")
            
        if user_win_nmass(user, computer):
            userwins()
            print(name, "WINS!!")
        
        else:
            return userloses()
            print("Computer WINS!!")
                
    else:
        print("Wrong Input!!")
        nmass_rules()
        #ontinue
                
    
    
if __name__ == '__main__':
    
    name = input("Enter your name: ")
    while True:
        print()
        print("Let's Play!!!")
        print("Which version of Hand Game?")
        print("Enter 1 to play Rock-Paper-Scissors")
        print("Enter 2 to play Rock-Paper-Scissors-Lizard-Spock")
        print("Enter 3 to play Nuke-Snake-Man-Sponge-Alien")
        print("Enter 4 to if you don't feel like playing.")
        print()


        try:
            choice = int(input("Enter your choice = "))
        except ValueError:
            print("Wrong Choice")
            continue

# Play the traditional version of the game
        if choice == 1:
            rps()

# Play the new version of the game
        elif choice == 2:
            rpsls()

# Play the new version of the game
        elif choice == 3:
            nmass()
            
# Quit the GAME LOOP 
        elif choice == 4:
            break

# Other wrong input
        else:
            print("Wrong choice. Read instructions carefully.")
    
    
```
